```python
passwords_df = pd.read_table("advent_calendar_d2.txt", header = None)
passwords_df
```

```python
with open("advent_calendar_d2.txt", "r", encoding="utf-8") as lst:
   passwords = lst.readlines()
```

```python
passwords
```

```python
correct_passwords = 0 

for line in passwords:
    #splitting the data into password, policy, desired frequency bounds, letter required in password
    entire_policy = line.split(':')[0]
    freq = entire_policy.split(" ")[0]
    freq_low = freq.split("-")[0]
    freq_high = freq.split("-")[1]
    letter = entire_policy.split(" ")[1]
    password = line.split(':')[1]
    
    #counting the occurences of the given letter in the password 
    count = 0 
    for char in password: 
        if char == letter:
            count += 1
    
    if int(freq_low) <= count <= int(freq_high):
        correct_passwords +=1 

print(correct_passwords)
```

```python
correct_passwords = 0 

for line in passwords:
    #splitting the data into password, policy, desired frequency bounds, letter required in password
    entire_policy = line.split(':')[0]
    freq = entire_policy.split(" ")[0]
    freq_low = freq.split("-")[0]
    freq_high = freq.split("-")[1]
    letter = entire_policy.split(" ")[1]
    password = line.split(':')[1]
    
    instances = 0
    for i, char in enumerate(password):
        if (i == int(freq_low) or i == int(freq_high)) and char == letter:
            instances += 1
    
    if instances == 1: 
        correct_passwords += 1

print(correct_passwords)
            
```
